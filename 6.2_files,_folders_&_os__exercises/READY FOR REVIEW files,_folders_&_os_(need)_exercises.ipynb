```python
month_key={
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}


text_split='oct-08_singapore.txt'.split('-')
print(month_key[text_split[0]]+"-"+(text_split[1]))
```

```python
def rename_my_file(old_file_name):
    text_split=old_file_name.split('-')
    if text_split[0].isnumeric() == False:
        return month_key[text_split[0]] + "-" + text_split[1]
    else:
        return old_file_name

print(rename_my_file('12-25_singapore.txt'),
rename_my_file('nov-25_singapore.txt'), sep="\n")
```

```python
import glob
import os
import shutil
import numpy as np
```

```python
new_file_names = []
old_file_paths=glob.glob('os-collaboration-exercise-data\\*.txt')

for old_file_path in old_file_paths:
    path_split = os.path.split(old_file_path)
    new_file_path=os.path.join(path_split[0],rename_my_file(path_split[1]))
    if os.path.exists(new_file_path) == False:
        os.rename(old_file_path,new_file_path)
```

```python
for city in cities:
    city_file_paths = glob.glob(f'os-collaboration-exercise-data/*{city}.txt')
    folder_file_path_list = []
    path_truth = []
    
    for city_file_path in city_file_paths:
        folder_file_path = os.path.join(os.path.split(city_file_path)[0],f'{city}',os.path.split(city_file_path)[1])
        folder_file_path_list += [folder_file_path] # forms a list of expected paths of the files in the folder
    path_truth = folder_file_path_list.copy()
    for index, folder_path in enumerate(folder_file_path_list):
        path_truth[index] = os.path.exists(folder_path) # changes folder_file_path_list into a True/False based on whether the path alr exists aka whether the file is already in the folder
    if all(path_truth): # if all files are already in the folder
        print(f"{city} folder complete")
    else:
        city_file_paths = np.array(city_file_paths)
        path_truth = np.array(path_truth)
        print(f"{city} folder incomplete")
        print(city_file_paths[~path_truth]," are not in their folder") # uses inverse mask to subset the paths of files that are not in their folders
        for file, folder_path in zip(city_file_paths, folder_file_path_list):
            shutil.move(file,folder_path)
            print(file, " moved to ", folder_path)
```
