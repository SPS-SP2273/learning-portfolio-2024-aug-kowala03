```python
def rename_my_file(old_file_name):
    month_key={
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }
    
    underscore_split = old_file_name.split('_')
    hyphen_split = underscore_split[0].split('-')
    
    if hyphen_split[0].isnumeric() == False:
        return month_key[hyphen_split[0]] + '-' + hyphen_split[1] + '-' + underscore_split[1]
    
    else:
        return old_file_name

print(rename_my_file('12-25-singapore.txt'),
rename_my_file('nov-25_singapore.txt'), sep="\n")
```

```python
import glob
import os
import shutil
import numpy as np
```

```python
new_file_names = []
old_file_paths = glob.glob('os-collaboration-exercise-data\\*.txt')

for old_file_path in old_file_paths:
    path_split = os.path.split(old_file_path)
    new_file_path = os.path.join(path_split[0], rename_my_file(path_split[1]))
    if os.path.exists(new_file_path) == False:
        os.rename(old_file_path, new_file_path)
    else:
        print('renamed file already exists')
```

```python
# 2nd attempt: replaces try except with if else, condition being whether the file exists in the 

for city in cities:
    city_file_paths = glob.glob(f'os-collaboration-exercise-data/*{city}.txt')
    
    for city_file_path in city_file_paths:
        path = os.path.join(os.path.split(city_file_path)[0],f'{city}',os.path.split(city_file_path)[1])
        
        if os.path.exists(path):
            print(f"{path} already exists, no need to move the file!")

        else:
            shutil.move(city_file_path, path)
            print(f'moved {city_file_path} to {path}')
```

```python
for city in cities:
    
    city_file_paths = glob.glob(f'os-collaboration-exercise-data/*{city}.txt')
    file_in_folder_pathList = []
    pathTruth = []
    
    for city_file_path in city_file_paths:
        city_file_pathSplit = os.path.split(city_file_path)
        file_in_folder_path = os.path.join(city_file_pathSplit[0],f'{city}',city_file_pathSplit[1])
        file_in_folder_pathList += [file_in_folder_path] # forms a list of expected paths of the files in the folder
    
    pathTruth = file_in_folder_pathList.copy()
    
    for index, path in enumerate(file_in_folder_pathList):
        pathTruth[index] = os.path.exists(path) # changes folder_file_path_list into a True/False based on whether the path alr exists aka whether the file is already in the folder
    
    if all(pathTruth): # if all files are already in the folder
        print(f"{city} folder complete")
    
    else:
        print(f"{city} folder incomplete")
        city_file_paths = np.array(city_file_paths)
        pathTruth = np.array(pathTruth)
        print(city_file_paths[~pathTruth], " are not in their folder") # use inverse mask to subset the paths of files that are not in their folders
        
        for file_path, file_path_in_folder in zip(city_file_paths, file_in_folder_pathList):
            shutil.move(file_path, file_path_in_folder)
            print(file_path, " moved to ", file_path_in_folder)
```
