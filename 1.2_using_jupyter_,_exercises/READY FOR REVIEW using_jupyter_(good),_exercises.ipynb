1<sup>st</sup> number: Mean. Average of time taken to run code in the cell. Shows how fast the code runs.

2<sup>nd</sup> number: Standard deviation. Represents the spread of data points from the mean. Statistically, 65% of runs will take time within 1 s.d. of the means, 95% within 2 s.d, 99.7% within 3 s.d.
