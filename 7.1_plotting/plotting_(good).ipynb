```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, num=100) # 100 points between -pi to pi
print(x)

y = np.linspace(-1,1,num=5) # evenly spaced
print(y)
```

```python
cos_x = np.cos(x)
sin_x = np.sin(x)
```

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
```

```python
# object oriented

fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')
ax[0].legend()
ax[1].legend()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=3, # creates 6 axis
                       figsize = (10,10), # 10 inch by 10 inch
                      sharex = True, sharey = True)

for a in ax.flatten(): # because if not a is a 2d list
    a.plot(x,sin_x)
    a.grid(alpha = 0.25)

plt.tight_layout() # stops overlap of labels
```

```python
fig, ax = plt.subplots(nrows=2,ncols=3, # creates 6 axis
                       figsize = (10,10), # 10 inch by 10 inch
                      sharex = 'col', sharey = 'row') # share x by col and y by row

for count, a in enumerate(ax.flatten()): # because if not a is a 2d list
    if count == 3:            # for the 4th axis,  different function is plotted
        a.plot(x, 10*cos_x)
        continue               # continue after count == 3
    
    a.plot(x,sin_x)
    a.grid(alpha = 0.25)

ax[-1,-1].set_xlim(-10,10) # for axis in last row last col, set x range, changes for 0,-1 also because they share x axis

plt.tight_layout() # stops overlap of labels
```

# 5 Multiple rows and columns

```python
x = np.linspace(-np.pi, np.pi, num = 100)

fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')           # define the layout

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten() # give names to the axis, usually the name of varaible you are plotting
# if dont use flatten, [[top_left, top_right], [bottom_left, bottom_right]] = ax

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)

x2 = np.arange(10)
y = x2**2

bottom_left.bar(x2,y)
bottom_right.plot(x, sin_x)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```
