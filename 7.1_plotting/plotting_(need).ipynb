```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = 2, 0.75

# Let's start plotting
for alpha in [0,0.25, 0.5, 0.75]:
    plt.plot(x, y_1, label='Y1 values',
              color='red', linestyle='dashed')
    plt.plot(x,y_2, label='Y2 values',c='blue')
    plt.errorbar(x,y_3, c='green', xerr=x_error, yerr=y_error)
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.title('X vs Y')
    plt.grid(f'{alpha=}')
    plt.legend(loc='upper left')
    plt.savefig('plot.png', dpi=150) # lossy, usually smaller
    plt.savefig('plot.jpeg', dpi=150)
    plt.savefig('plot.pdf', dpi=150) # still very clear upon zooming in as it stores info as vectors, formula to draw the obj)
# bc plot isnt cleared, plt adds a plot each time it's run, end up with 3*3+1
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = 2, 0.75

# Let's start plotting
for alpha in [0,0.25, 0.5, 0.75]:
    plt.plot(x, y_1, label='Y1 values',
              color='red', linestyle='dashed')
    plt.plot(x,y_2, label='Y2 values',c='blue')
    plt.errorbar(x,y_3, c='green', xerr=x_error, yerr=y_error)
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.title('X vs Y')
    plt.grid(f'{alpha=}')
    plt.legend(loc='upper left')
    plt.savefig('plot.png', dpi=150) # lossy, usually smaller
    plt.savefig('plot.jpeg', dpi=150)
    plt.savefig('plot.pdf', dpi=150) # still very clear upon zooming in as it stores info as vectors, formula to draw the obj)
    plt.show()
    plt.clf()
```

```python
x = np.linspace(0,20)
x_2=x**2
sin_x = np.sin(x)

plt.plot(x_2, sin_x,color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 0.5)

# peaks a bit jagged, not smooth
```

```python
x = np.linspace(0,6, 1000) # change the number of points used
x_2=x**2
sin_x = np.sin(x)

plt.plot(x,x_2,color='red', linestyle='dashed',label='x^2')
plt.legend(loc='lower left')
plt.grid(alpha = 0.5)

plt.twinx() # adds new y 

plt.plot(x,sin_x,color='blue', linestyle='solid',label='sin(x)')
plt.legend(loc='lower right')

plt.xlabel('x-values')
plt.ylabel('y-values')

plt.title('X vs Y')
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
file_url='https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Land_and_Ocean_summary.txt'
data = np.loadtxt(file_url, skiprows=23)

date = data[:,0]
anomaly = data[:,-1]

plt.scatter(date, anomaly, alpha=.5)
```

```python
url='https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Complete_TAVG_daily.txt'
data = np.loadtxt(url,
                   skiprows=24) # skipping rows of intro text

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0,date[0],date[-1],linestyle='--',color='gray')
plt.ylim([-4, 4])
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight')

url='https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Complete_TAVG_daily.txt'
data = np.loadtxt(url,
                   skiprows=24) # skipping rows of intro text

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0,date[0],date[-1],linestyle='--',color='gray')
plt.ylim([-4, 4])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temparature Anomaly\n(Relative to average from Jan 1951 - Dec 1980.)')
```
